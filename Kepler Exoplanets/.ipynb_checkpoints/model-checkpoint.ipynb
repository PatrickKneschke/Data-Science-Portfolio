# Problem
### Objective
Given the measurements of te Kepler mission the goal is to determine if a Kepler Object of Interest (KOI) is likely to be confirmed as an exoplanet or rejected as a false positive.
### Method
The data entries marked as "Confirmed" and "False Positive" are selected, discarding all entries marked as "Candidate". Heavily scewed features are log transformed and missing values filled with the median value and minmax-scaling applied.
The koi_score is removed from the data and instead used for a reference model to compare results to. Three classifiers will be trained (Logistic Regression, Decicion Tree ans Random Forest) in three stages :
1. Excluding features representing the uncertainties of Kepler measurements and certain features with low correlation to the target
2. The full set of features
3. A reduced set of features based on feature importances from the decision tree in stage 2
### Bonus
Using the best model a prediction will be made for the remaining data entries, marked as "Candidate".

# Libraries

```python
#misc
import warnings
warnings.filterwarnings('ignore')

# data
import pandas as pd
import numpy as np

# preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# modeling
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

#plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style( "whitegrid")
```

# Utility functions

```python
def make_confusion_matrix(cfm, ax=None, group_names=None, categories='auto', count=True, percent=True,
                          cbar=True, xyticks=True, xyplotlabels=True, sum_stats=True,
                          figsize=None, cmap='Blues', title=None) :
    '''
    This function will make a pretty plot of an sklearn Confusion Matrix cm using a Seaborn heatmap visualization.

    Arguments
    ---------
    cfm:           confusion matrix to be passed in

    group_names:   List of strings that represent the labels row by row to be shown in each square.

    categories:    List of strings containing the categories to be displayed on the x,y axis. Default is 'auto'

    count:         If True, show the raw number in the confusion matrix. Default is True.

    normalize:     If True, show the proportions for each category. Default is True.

    cbar:          If True, show the color bar. The cbar values are based off the values in the confusion matrix.
                   Default is True.

    xyticks:       If True, show x and y ticks. Default is True.

    xyplotlabels:  If True, show 'True Label' and 'Predicted Label' on the figure. Default is True.

    sum_stats:     If True, display summary statistics below the figure. Default is True.

    figsize:       Tuple representing the figure size. Default will be the matplotlib rcParams value.

    cmap:          Colormap of the values displayed from matplotlib.pyplot.cm. Default is 'Blues'
                   See http://matplotlib.org/examples/color/colormaps_reference.html
                   
    title:         Title for the heatmap. Default is None.

    '''

    # CODE TO GENERATE TEXT INSIDE EACH SQUARE
    blanks = ["" for i in range( cfm.size)]
    if group_names and len( group_names) == cfm.size :
        group_labels = ["{}\n".format( value) for value in group_names]
    else :
        group_labels = blanks

    if count :
        group_counts = ["{0:0.0f}\n".format( value) for value in cfm.flatten()]
    else :
        group_counts = blanks

    if percent :
        group_percentages = ["{0:.2%}".format( value) for value in cfm.flatten() / np.sum( cfm)]
    else :
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip( group_labels, group_counts, group_percentages)]
    box_labels = np.asarray( box_labels).reshape( cfm.shape[0], cfm.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats :
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace( cfm) / float( np.sum( cfm))

        #if it is a binary confusion matrix, show some more stats
        if len( cfm) == 2 :
            #Metrics for Binary Confusion Matrices
            precision = cfm[1,1] / sum( cfm[:,1])
            recall    = cfm[1,1] / sum( cfm[1,:])
            f1_score  = 2 * precision * recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy, precision, recall, f1_score
            )
        else :
            stats_text = "\n\nAccuracy={:0.3f}".format( accuracy)
    else :
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize == None :
        #Get default figure size if not set
        figsize = plt.rcParams.get( 'figure.figsize')

    if xyticks == False :
        #Do not show categories if xyticks is False
        categories = False


    # MAKE THE HEATMAP VISUALIZATION
    if ax == None :
        plt.figure( figsize=figsize)
    sns.heatmap( cfm, annot=box_labels, fmt="", cmap=cmap, cbar=cbar, xticklabels=categories, yticklabels=categories, ax=ax)

    if xyplotlabels :
        if ax == None :
            plt.ylabel( 'True label')
            plt.xlabel( 'Predicted label' + stats_text)
        else :
            ax.set_ylabel( 'True label')
            ax.set_xlabel( 'Predicted label' + stats_text)
    else :
        if ax == None :
            plt.xlabel( stats_text)
        else :
            ax.set_xlabel( stats_text)
    
    if title :
        if ax == None :
            plt.title( title)
        else :
            ax.set_title( title)
```

# Load Data

```python
data = pd.read_csv( "data/cumulative.csv")
data.head()
```

# Feature Selection

```python
# core features to use in first round
features = [
    #"koi_score", 
    "koi_fpflag_nt", "koi_fpflag_ss", "koi_fpflag_co", "koi_fpflag_ec",
    "koi_period",
    "koi_time0bk",
    "koi_impact",
    "koi_duration",
    "koi_depth",
    "koi_prad",
    "koi_teq",
    "koi_steff",
    "koi_srad",
    "ra",
    "dec",
]

# additional features of less importance, to check if they improve model performance in second round
features_ext = [
    "koi_period_err1", "koi_period_err2",
    "koi_time0bk_err1", "koi_time0bk_err2",
    "koi_impact_err1", "koi_impact_err2",
    "koi_duration_err1", "koi_duration_err2",
    "koi_depth_err1", "koi_depth_err2",
    "koi_prad_err1", "koi_prad_err2",
    "koi_steff_err1", "koi_steff_err2",
    "koi_srad_err1", "koi_srad_err2",
    "koi_model_snr",
    "koi_kepmag"
]

# select all relevant features and target
df = data[features + features_ext + ["koi_disposition"]]

# use only "CONFORMED" and "FALSE POSITIVE" koi
df = df[ df["koi_disposition"] != "CANDIDATE"].replace(to_replace=["FALSE POSITIVE", "CONFIRMED"], value=[0, 1])
df.head()
```

# Data Processing

```python
#  scewed features to log transform
cols_to_log = [
    'koi_period',
    'koi_time0bk', 
    'koi_impact',
    'koi_duration',
    'koi_depth',
    'koi_prad',
    'koi_srad',
    'koi_teq',
    "koi_period_err1", "koi_period_err2",
    "koi_time0bk_err1", "koi_time0bk_err2",
    "koi_impact_err1", "koi_impact_err2",
    "koi_duration_err1", "koi_duration_err2",
    "koi_depth_err1", "koi_depth_err2",
    "koi_prad_err1", "koi_prad_err2",
    "koi_srad_err1", "koi_srad_err2",
    "koi_model_snr",
]
remaining_cols = [x for x in df.columns if x not in cols_to_log]

def log1pabs( x) :
    return np.log1p( np.abs( x))

log_trafo = ColumnTransformer(
    transformers=[("log_trafo", FunctionTransformer( log1pabs), cols_to_log)],
    remainder="passthrough"
)
imputer = SimpleImputer( strategy="median")
minmax_scaler = MinMaxScaler()

preprocess_pipeline =  Pipeline(
    steps=[
        ('log_trafo', log_trafo), 
        ('imputer', imputer),
        ('minmax_scaler', minmax_scaler),
    ]
)
```

```python
df2 = pd.DataFrame( preprocess_pipeline.fit_transform( df), columns=(cols_to_log + remaining_cols))
df2
```

# Models

```python
X = df2[features]
Xext = df2[features + features_ext]
Xref = data[data["koi_disposition"] != "CANDIDATE"].replace( to_replace=["FALSE POSITIVE", "CONFIRMED"], value=[0, 1])
Xref["koi_score"] = Xref["koi_score"].fillna( Xref["koi_score"].median())
y = df2["koi_disposition"]

X_train, X_test, Xext_train, Xext_test, Xref_train, Xref_test, y_train, y_test = train_test_split( X, Xext, Xref, y, test_size=0.2, random_state=123)
```

### Reference model based only on "koi_score"

```python
class ReferenceClf( ClassifierMixin, BaseEstimator) :
    def fit(self, X, y=None):
        self.is_fitted_ = True
        return self
        
    def predict(self, X):
        return (X["koi_score"] > 0.5).astype(np.int32)
    
```

```python
refclf = ReferenceClf()
print( "reference model accuracy : ", refclf.score( Xref_test, y_test))
```

### Logistic Regression

```python
params = {
    "penalty" : ["l1", "l2"],  
    "C" : [0.01, 0.1, 1.0], 
    "solver" : ["lbfgs", "liblinear", "newton-cg", "saga"],
}

logreg = GridSearchCV( LogisticRegression(), params, cv=10)
logreg.fit( X_train, y_train)
print( "best params : ", logreg.best_params_, " , best score : ", logreg.best_score_)
print( "logistic regression accuracy : ", logreg.score( X_test, y_test))
```

```python
logreg_ext = GridSearchCV( LogisticRegression(), params, cv=10) 
logreg_ext.fit( Xext_train, y_train)
print( "best params : ", logreg_ext.best_params_, " , best score : ", logreg_ext.best_score_)
print( "logistic regression accuracy : ", logreg_ext.score( Xext_test, y_test))
```

### Decision Tree

```python
params = {
    "criterion" : ["gini", "entropy", "log_loss"],  
    "max_depth" : [None, 5, 10, 20], 
    "max_features" : ["sqrt", "log2"],
}

tree = GridSearchCV( DecisionTreeClassifier(), params, cv=10)
tree.fit( X_train, y_train)
print( "best params : ", tree.best_params_, " , best score : ", tree.best_score_)
print( "decicion tree accuracy : ", tree.score( X_test, y_test))
```

```python
tree_ext = GridSearchCV( DecisionTreeClassifier(), params, cv=10)
tree_ext.fit( Xext_train, y_train)
print( "best params : ", tree_ext.best_params_, " , best score : ", tree_ext.best_score_)
print( "decision tree accuracy : ", tree_ext.score( Xext_test, y_test))
```

```python
importance = pd.DataFrame( tree_ext.best_estimator_.feature_importances_, index=Xext_train.columns, columns=["importance"]).sort_values(by="importance", ascending=False)
importance.plot( kind="bar")
```

### Random Forest

```python
params = {
    "n_estimators" : [10, 100, 1000],
    "criterion" : ["entropy"],  
    "max_depth" : [None, 5, 10], 
    "max_features" : ["sqrt"],
}

forest = GridSearchCV( RandomForestClassifier(), params, cv=10)
forest.fit( X_train, y_train)
print( "best params : ", forest.best_params_, " , best score : ", forest.best_score_)
print( "random forest accuracy : ", forest.score( X_test, y_test))
```

```python
forest_ext = GridSearchCV( RandomForestClassifier(), params, cv=10)
forest_ext.fit( Xext_train, y_train)
print( "best params : ", forest_ext.best_params_, " , best score : ", forest_ext.best_score_)
print( "random forest accuracy : ", forest_ext.score( Xext_test, y_test))
```

## Reduced set of features based on feature importance

```python
# use features with importances above 1 percent
features_red = importance[ importance["importance"] >= 0.01].index.to_list()

Xred = df2[features_red]
Xred_train, Xred_test = train_test_split( Xred, test_size=0.2, random_state=123)

# log reg
logreg_red = LogisticRegression( C=1.0, penalty="l1", solver="liblinear")
logreg_red.fit( Xred_train, y_train)
print( "logistic regression accuracy : ", logreg_red.score( Xred_test, y_test))

# tree
tree_red = DecisionTreeClassifier( criterion="entropy", max_depth=None, max_features="sqrt")
tree_red.fit( Xred_train, y_train)
print( "decision tree accuracy : ", tree_red.score( Xred_test, y_test))

# random forest
forest_red = RandomForestClassifier( criterion="entropy", max_depth=None, max_features="sqrt", n_estimators=100)
forest_red.fit( Xred_train, y_train)
print( "random forest accuracy : ", forest_red.score( Xred_test, y_test))
```

# Conclusion
The disposition of Kepler Objects of Interests (KOI) have been predicted using various classifiers. The prediction has been performed using only the direct measurements of the Kepler mission as input, excluding the "koi_score", which is a probability assigned to each KOI by Keplers own Robovetter. The koi_score is instead used as a benchmark to compare model quality.

The classifiers that have been trained are Logistic Regression, Decision Tree and a Random Forest, each trained and optimized using k-fold cross-validation to find an ideal parameter configuration. Modeling has been done in three stages, starting with a reduced set of features, excluding the uncertainties of each measurement as well as some features with small correlation to the dispostion, followed by a second run using all features.
Finally another set of models is trained using the features with the hightest importance according to the decision tree trained in the second stage.

```python
results_df = pd.DataFrame(
    [
        [logreg.score(X_test, y_test), logreg_ext.score(Xext_test, y_test), logreg_red.score(Xred_test, y_test)],
        [tree.score(X_test, y_test), tree_ext.score(Xext_test, y_test), tree_red.score(Xred_test, y_test)],
        [forest.score(X_test, y_test), forest_ext.score(Xext_test, y_test), forest_red.score(Xred_test, y_test)],
    ],
    index=["LogReg", "Tree", "Forest"],
    columns=["core features", "extendex features", "reduced features"],
)

results_df
```

```python
fig, ax = plt.subplots( 1, 4, figsize=(24, 5))

cfm1 = confusion_matrix( y_test, logreg_red.predict( Xred_test))
cfm2 = confusion_matrix( y_test, tree_red.predict( Xred_test))
cfm3 = confusion_matrix( y_test, forest_red.predict( Xred_test))
cfm4 = confusion_matrix( y_test, refclf.predict( Xref_test))

make_confusion_matrix(cfm1, title="LogReg", ax=ax[0])
make_confusion_matrix(cfm2, title="Tree", ax=ax[1])
make_confusion_matrix(cfm3, title="Forest", ax=ax[2])
make_confusion_matrix(cfm4, title="Reference", ax=ax[3])
```

Overall the random forest shows the best performance, approching even the model based on koi_score. The reduced set of features based on feature importances slightly improved the model accurary of logistic regression and the decision tree, but the random forest displayed similarly high accuracy in all three stages.
